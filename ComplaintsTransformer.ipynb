```python
import pandas as pd
import numpy as np
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
#!pip install simpletransformers
from simpletransformers.classification import ClassificationModel
import re
```

```python
df = pd.read_csv('transformercomplaints.csv', nrows = 30000)
df.head()
```

```python
df.shape
```

```python
# Create descriptions and label lists
descriptions = df['complaint'].tolist()
labels = df['issue'].tolist()
```

```python
# Split data into train and test sets
train_descriptions, test_descriptions, train_labels, test_labels = train_test_split(descriptions, labels, test_size=0.3)
```

```python
# Encode the labels as numerical values
label_dict = {label: i for i, label in enumerate(set(labels))}
train_labels = [label_dict[label] for label in train_labels]
test_labels = [label_dict[label] for label in test_labels]
```

```python
# create the train and test dataframes
train_df = pd.DataFrame({'text': train_descriptions, 'labels': train_labels})
test_df = pd.DataFrame({'text': test_descriptions, 'labels': test_labels})
train_df.labels.value_counts()
```

```python
model_args = {
    'train_batch_size': 16,
    'eval_batch_size': 16,
    'max_seq_length': 512,
    'num_train_epochs': 5,
    'overwrite_output_dir': True,
    'evaluate_during_training': False,
}
model = ClassificationModel('distilbert', 'distilbert-base-uncased', num_labels=len(label_dict), args=model_args, use_cuda=True)
model.train_model(train_df, eval_df=test_df)
```

Running loss per epoch
1. Epochs 0/5. Running Loss: 2.3303
2. Epochs 1/5. Running Loss: 1.7604
3. Epochs 2/5. Running Loss: 0.1220
4. Epochs 3/5. Running Loss: 0.6659
5. Epochs 4/5. Running Loss: 0.6450

```python
# Save the model
model.save_model('distilbert')
```

```python
# Evaluate the model
result, model_outputs, wrong_predictions = model.eval_model(test_df)
```

```python
# Model Accuracy
Accuracy = 100 - (len(wrong_predictions)/len(test_df)*100)
print(Accuracy)
```
